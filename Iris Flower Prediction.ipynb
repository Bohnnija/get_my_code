```python
#Importing the important libraries

import numpy as np
import matplotlib.pyplot as plt #to plot graphs
import seaborn as sns #graph model
import pandas as pd
import os
%matplotlib inline

from sklearn import datasets

iris = datasets.load_iris()
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df = df.drop('Id',axis = 1)
df

#Id column is dropped because it was adding no value to our data for exploration
```

```python
x = df.drop('Species',axis = 1)
y = df['Species']
#Here we divided the data frame as input and output respectively as x and y
```

```python
from sklearn.model_selection import train_test_split
#we imported train test split which will divide our dataframe into 2 parts that
#is one part for training and the other for testing
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)

#Now we have our training and testing datasets ready, Training data set are x_train and y_train and testing data set is x_test and the sample we took for testing is 30%
#of our original data set df
```

```python
from sklearn.linear_model import LogisticRegression
#we imported logistic Regression
```

```python
ref = LogisticRegression()
#we are creating a object ref for the class Logisic Regression as above
```

```python
ref.fit(x_train,y_train)

#The above step executed the training part and now our model is trained and is ready to make predictions for any new input
#The above step executed the training part and now our model is  trained and is ready to make predictions for any new input
```

```python
ref.predict(x_test)
#we are testing our model against a testing data,x_test
```

```python
ref.score(x_test,y_test)
#The above code shows out model's accuract on the testing set
```

```python
ref.score(x_train,y_train)
```

```python
#his is the accuracy of our model on the testing set, this is done to ensure that our model does not overfit, and
#by the above observation it is clear (train score and test score are close to each other) that our model has
#low bias and low variance and is a reliable model.
```
